### Why we need Notation in mathematics?

Notation in mathematics is used to represent mathematical concepts and operations in a concise, clear, and standardized way. It helps to communicate mathematical ideas effectively and avoid ambiguity. Here are some of the reasons why notation is important in mathematics:

- **Clarity:** Notation allows mathematical expressions to be written in a compact and readable form, making it easier to understand and interpret mathematical concepts.

- **Consistency:** Notation helps ensure that mathematical ideas are expressed consistently across different texts, making it easier to compare and learn from different sources.

- **Precision:** Notation provides a precise way to express mathematical concepts, reducing the chances of misunderstanding and allowing for more accurate communication of mathematical ideas.

- **Ease of Calculation:** Notation makes it easier to perform mathematical operations, as symbols and shorthand can often simplify complex calculations.

- **Computation:** In the age of computers and automation, notation plays a crucial role in mathematical computation and data analysis. Mathematical notation is used as the input for mathematical software and computational tools, making it an essential tool in modern mathematics.

Overall, notation in mathematics is a powerful tool for effectively communicating mathematical ideas and solving mathematical problems.

```python
import numpy as np

# Input data
X = np.array([[2.5, 2.4], [0.5, 0.7], [2.2, 2.9], [1.9, 2.2], [3.1, 3.0],
              [2.3, 2.7], [2, 1.6], [1, 1.1], [1.5, 1.6], [1.1, 0.9]])

# Mean normalization
X = X - np.mean(X, axis=0)

# Calculate the covariance matrix
cov_matrix = np.cov(X.T)

# Calculate the eigenvalues and eigenvectors of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Select the top k eigenvectors (k is the number of dimensions you want to reduce to)
top_k_eigenvectors = eigenvectors[:, :2]

# Project the data onto the new reduced dimensional space
reduced_data = np.dot(X, top_k_eigenvectors)
reduced_data
```

```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
%matplotlib notebook

# Define the symbol x
x = sympy.Symbol('x')

# Define the function f(x)
f = x**3 + x**2 + x + 1

# Calculate the derivative of f(x)
df = f.diff(x)

# Evaluate the original function and its derivative for a range of x values
x_vals = np.linspace(-2, 2, 100)
y_vals = [f.evalf(subs={x: xi}) for xi in x_vals]
dy_vals = [df.evalf(subs={x: xi}) for xi in x_vals]

# Plot the original function and its derivative
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, dy_vals, label="f'(x)")
plt.legend()
plt.grid()
plt.axvline(x=0, color='black')
plt.axhline(y=0, color='black')

plt.show()

```

```python
import numpy as np
import sympy
import matplotlib.pyplot as plt

# Define the symbol x
x = sympy.Symbol('x')

# Define the function f(x)
f = x**3 + x**2 + x + 1

# Calculate the derivative of f(x)
df = f.diff(x)

# Evaluate the original function and its derivative for a range of x values
x_vals = np.linspace(-2, 2, 100)
y_vals = [f.evalf(subs={x: xi}) for xi in x_vals]
dy_vals = [df.evalf(subs={x: xi}) for xi in x_vals]

# Find the slope at x=0
x0 = 3
slope = df.evalf(subs={x: x0})

# Plot the original function and its derivative
plt.plot(x_vals, y_vals, label='f(x)')
plt.plot(x_vals, dy_vals, label="f'(x)")

# Add a vertical line and label for the x=0 point
plt.axvline(x0, color='red', linestyle='--')
plt.text(x0, -10, f"x={x0}, f'(x)={slope:.2f}", rotation=90, va='bottom')

plt.legend()
plt.show()
```
